# Fazendo o Import dos dados

Nesta etapa, defina o arquivo PDF que será processado. Para isso, utilizei a biblioteca `tabula` , que permite extrair tabelas diretamente de arquivos PDF.

Após realizar a leitura do conteúdo, exiba o resultado inicial para verificar como os dados foram importados. Observe que a tabela original é organizada por dias da semana , o que dificulta a análise cronológica das informações.

```python
import tabula

pdf_path = '/content/Central do Aluno.pdf'

dfs = tabula.read_pdf(pdf_path,pages = 'all' ,stream = True)

print(len(dfs))
dfs[0]
```

#Organizando os Dados

Neste trecho, o foco está na manipulação da tabela extraída do PDF para ordenar os dados pela coluna "DADOS FINAIS".

Foi utilizada a biblioteca `pandas` para realizar a manipulação e ordenação dos dados. Por padrão, o `pandas` trabalha com dados no formato ano-mês-dia, por isso foi necessário convertê-los para o padrão brasileiro ( dia-mês-ano).

Após a ordenação, os dados organizados foram exportados para um arquivo CSV para facilitar o acesso e reutilização.

```python
import pandas as pd
df = dfs[0]
df['DATA FINAL'] = pd.to_datetime(df['DATA FINAL'], format='%d/%m/%Y')
df_sorted = df.sort_values(by='DATA FINAL')
df_sorted['DATA FINAL'] = df_sorted['DATA FINAL'].dt.strftime('%d/%m/%Y')

df_sorted.to_csv('/content/horarios_ordenados.csv', index=False)

df_sorted
```
