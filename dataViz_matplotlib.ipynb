# üìä An√°lise de Vendas nas Lojas A, B, C e D

Este projeto gera visualiza√ß√µes das vendas mensais das lojas **A, B, C e D** ao longo do ano de 2022, utilizando **Matplotlib** para criar gr√°ficos comparativos.

## üìù Descri√ß√£o

O c√≥digo cria uma **figura com 4 subplots** (2x2), onde cada gr√°fico representa a evolu√ß√£o das vendas de uma loja ao longo do tempo.  
Algumas melhorias foram implementadas para uma melhor visualiza√ß√£o dos dados:
- Ajuste de espa√ßamentos entre os subplots (`wspace` e `hspace`).
- Remo√ß√£o de bordas superiores e direitas para um design mais limpo.
- Personaliza√ß√£o da grade (`grid`) para facilitar a leitura dos gr√°ficos.
- Formata√ß√£o do eixo X para exibir os meses corretamente.
- Aplica√ß√£o de legenda para cada gr√°fico.

## Set dos dados para  an√°lise

````python
import pandas as pd
import matplotlib.pyplot as plt

lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

# Criando DataFrame
df = pd.DataFrame(vendas_2022, index=lojas)
````

````python
df.head()
````

## Criando gr√°fico com subplots 2x2.

````python
import matplotlib.pyplot as plt


fig, axs = plt.subplots(2, 2, figsize=(14, 8))

# Ajustar os espa√ßamentos entre os subplots
plt.subplots_adjust(wspace=0.3, hspace=0.4)

# Adicionando um t√≠tulo geral
fig.suptitle('Vendas no per√≠odo de janeiro a dezembro de 2022 nas lojas A, B, C e D', fontsize=16)

# Adicionar os gr√°ficos em cada subplot
lojas = ['A', 'B', 'C', 'D']
for i, ax in enumerate(axs.flat):
    loja = lojas[i]
    ax.plot(df.loc[loja], label=f'Loja {loja}', linewidth=2)
    ax.set_title(f'Vendas na loja {loja}')
    ax.set_xlabel('M√™s')
    ax.set_ylabel('N√∫mero de vendas')
    ax.legend()

    # Ajustando o Y
    ax.set_ylim(50, 400)
    ax.yaxis.set_tick_params(labelsize=12)
    ax.xaxis.set_tick_params(labelsize=12)

    # Est√©tica
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(True, linestyle="--", alpha=0.6)

plt.show()
````
