## Set dos dados para  an치lise

````python
import pandas as pd
import matplotlib.pyplot as plt

lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

# Criando DataFrame
df = pd.DataFrame(vendas_2022, index=lojas)
````

````python
df.head()
````

## Criando gr치fico com subplots 2x2.

````python
import matplotlib.pyplot as plt

cores = ['darkviolet', 'green', 'darkblue', 'coral']

fig, axs = plt.subplots(2, 2, figsize=(14, 8))

# Ajustar os espa칞amentos entre os subplots
plt.subplots_adjust(wspace=0.3, hspace=0.4)

# Adicionando um t칤tulo geral
fig.suptitle('Vendas no per칤odo de janeiro a dezembro de 2022 nas lojas A, B, C e D', fontsize=16)

#Loop para plotar e customizar os subplots
for i, ax in enumerate(axs.flat):
    ax.plot(df.loc[df.index[i]], color=cores[i], lw=3)
    ax.set_title(f'Vendas na loja {df.index[i]}', loc='left', fontsize=16)
    ax.set_xlabel('M칡s', fontsize=14)
    ax.set_ylabel('N칰mero de vendas', fontsize=14)
    ax.tick_params(labelsize=12)
    ax.grid(color='lightgrey')

    # Ajustando o Y
    ax.set_ylim(50, 400)
    ax.yaxis.set_tick_params(labelsize=12)
    ax.xaxis.set_tick_params(labelsize=12)

    # Est칠tica
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(True, linestyle="--", alpha=0.6)

plt.show()
````
